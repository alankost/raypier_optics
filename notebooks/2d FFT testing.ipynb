```python
import numpy
from numpy import fft
from matplotlib import pyplot as pp
```

```python

x_ = numpy.linspace(-10,10,1024)
y_ = numpy.linspace(-10,10,1024)

x, y = numpy.meshgrid(x_,y_)

E_field = ((x**2 + y**2) < (0.5**2)).astype(numpy.double)
```

```python
pp.imshow(E_field)
```

```python
out = fft.fft2(fft.fftshift(E_field))
```

```python
shifted = fft.fftshift(out)
```

```python
pwr = shifted.real**2 + shifted.imag**2
```

```python
pp.imshow(numpy.log(pwr[400:600,400:600]))
```

```python
pp.imshow(shifted[450:550,450:550].real)
```
